```python
# ═══════════════════════════════════════════════════════════════════
# SECTION 2: DATA LOADING
# ═══════════════════════════════════════════════════════════════════

# Upload files to Colab
from google.colab import files

print("Please upload your 4 CSV files:")
print("1. Applicant Data CSV")
print("2. SEVIS Data CSV")
print("3. Connect Data CSV")
print("4. Outreach Data CSV")
print()

uploaded = files.upload()

# ───────────────────────────────────────────────────────────────────
# Automatically detect uploaded filenames
# ───────────────────────────────────────────────────────────────────

print("\nDetecting uploaded files...")
print("─" * 70)

uploaded_files = list(uploaded.keys())
print(f"Found {len(uploaded_files)} files:")
for i, filename in enumerate(uploaded_files, 1):
    print(f"  {i}. {filename}")

# ───────────────────────────────────────────────────────────────────
# TODO: Match uploaded files to dataset types
# Adjust the keywords if your filenames are different
# ───────────────────────────────────────────────────────────────────

def find_file(keyword, uploaded_files):
    """Find file containing keyword in filename"""
    for filename in uploaded_files:
        if keyword.lower() in filename.lower():
            return filename
    return None

# Auto-detect files based on keywords
APPLICANT_FILE = find_file('Applicant', uploaded_files)
SEVIS_FILE = find_file('SEVIS', uploaded_files)
CONNECT_FILE = find_file('Connect', uploaded_files)
OUTREACH_FILE = find_file('Outreach', uploaded_files)

print("\nFile mapping:")
print(f"  Applicant → {APPLICANT_FILE}")
print(f"  SEVIS     → {SEVIS_FILE}")
print(f"  Connect   → {CONNECT_FILE}")
print(f"  Outreach  → {OUTREACH_FILE}")
print()

# ───────────────────────────────────────────────────────────────────
# Load datasets with error handling
# ───────────────────────────────────────────────────────────────────

def load_dataset(filename, dataset_name):
    """
    Load CSV file with proper error handling

    Parameters:
    -----------
    filename : str
        Name of the CSV file
    dataset_name : str
        Descriptive name for logging

    Returns:
    --------
    pd.DataFrame or None
    """
    if filename is None:
        print(f"✗ {dataset_name}: File not found in uploads")
        return None

    try:
        df = pd.read_csv(filename, low_memory=False)
        print(f"✓ Loaded {dataset_name}: {df.shape[0]:,} rows × {df.shape[1]} columns")
        return df
    except Exception as e:
        print(f"✗ Error loading {dataset_name}: {str(e)}")
        return None

# Load all datasets
print("Loading datasets...")
print("─" * 70)

applicant_df = load_dataset(APPLICANT_FILE, 'Applicant Data')
sevis_df = load_dataset(SEVIS_FILE, 'SEVIS Data')
connect_df = load_dataset(CONNECT_FILE, 'Connect Data')
outreach_df = load_dataset(OUTREACH_FILE, 'Outreach Data')

print("─" * 70)

# Verify all loaded successfully
if all([applicant_df is not None, sevis_df is not None,
        connect_df is not None, outreach_df is not None]):
    print("✓ All datasets loaded successfully\n")
else:
    print("⚠ Some datasets failed to load. Please check filenames.\n")
```

```python
# ═══════════════════════════════════════════════════════════════════
# SECTION 3: INITIAL INSPECTION
# ═══════════════════════════════════════════════════════════════════

print("DATASET OVERVIEW")
print("═" * 70)

# ───────────────────────────────────────────────────────────────────
# TODO: Create a summary DataFrame showing key metrics for each dataset
# Hint: Include dataset name, rows, columns, memory usage
# ───────────────────────────────────────────────────────────────────

datasets = {
    'Applicant': applicant_df,
    'SEVIS': sevis_df,
    'Connect': connect_df,
    'Outreach': outreach_df
}

overview_data = []
for name, df in datasets.items():
    overview_data.append({
        'Dataset': name,
        'Rows': f"{df.shape[0]:,}",
        'Columns': df.shape[1],
        'Memory_MB': round(df.memory_usage(deep=True).sum() / 1024**2, 2)
    })

overview_df = pd.DataFrame(overview_data)
print(overview_df.to_string(index=False))
print()

# ───────────────────────────────────────────────────────────────────
# Identify relationship keys (KNOT FILE concept)
# ───────────────────────────────────────────────────────────────────

print("\nRELATIONSHIP KEY ANALYSIS")
print("═" * 70)

# TODO: Check which datasets have Reference_ID and SEVIS_ID
# This identifies the "KNOT FILE" that connects all datasets

relationship_data = []
for name, df in datasets.items():
    has_ref = 'Reference_ID' in df.columns
    has_sevis = 'SEVIS_ID' in df.columns

    # Determine role
    if has_ref and has_sevis:
        role = '🔑 KNOT FILE (Has both IDs)'
    elif has_ref:
        role = 'Links via Reference_ID'
    elif has_sevis:
        role = 'Links via SEVIS_ID'
    else:
        role = 'No clear link key'

    relationship_data.append({
        'Dataset': name,
        'Has_Reference_ID': '✓' if has_ref else '✗',
        'Has_SEVIS_ID': '✓' if has_sevis else '✗',
        'Role': role
    })

relationship_df = pd.DataFrame(relationship_data)
print(relationship_df.to_string(index=False))
print("\n💡 Insight: Applicant is the KNOT FILE - it can join to all other datasets!\n")
```

```python
# ═══════════════════════════════════════════════════════════════════
# SECTION 6: MISSING VALUE ANALYSIS
# ═══════════════════════════════════════════════════════════════════

print("MISSING VALUE ANALYSIS")
print("═" * 70)

# ───────────────────────────────────────────────────────────────────
# TODO: Calculate missing value statistics for each dataset
# Show columns with >10% missing data
# ───────────────────────────────────────────────────────────────────

def analyze_missing_values(df, dataset_name, threshold=10):
    """
    Analyze missing values in a dataset

    Parameters:
    -----------
    df : pd.DataFrame
        Dataset to analyze
    dataset_name : str
        Name of dataset for display
    threshold : int
        Only show columns with missing % above this threshold

    Returns:
    --------
    pd.DataFrame with missing value statistics
    """
    missing_data = []

    for col in df.columns:
        null_count = df[col].isna().sum()
        null_pct = (null_count / len(df)) * 100

        if null_pct > threshold:
            missing_data.append({
                'Column': col,
                'Missing_Count': null_count,
                'Missing_%': round(null_pct, 2),
                'Total_Rows': len(df)
            })

    if missing_data:
        missing_df = pd.DataFrame(missing_data).sort_values('Missing_%', ascending=False)
        print(f"\n{dataset_name}: Columns with >{threshold}% missing data")
        print("─" * 70)
        print(missing_df.to_string(index=False))
        return missing_df
    else:
        print(f"\n{dataset_name}: ✓ No columns with >{threshold}% missing data")
        return pd.DataFrame()

# Analyze each dataset
all_missing = []

for name, df in datasets.items():
    missing_df = analyze_missing_values(df, name, threshold=10)
    if not missing_df.empty:
        missing_df['Dataset'] = name
        all_missing.append(missing_df)

print("\n" + "═" * 70)
print("⚠ CRITICAL FINDING: Country field in Applicant has 99.7% missing!")
print("  This blocks all geographic analysis and agency performance by region.")
print("═" * 70)
```

```python
# ═══════════════════════════════════════════════════════════════════
# SECTION 7: DETAILED COLUMN PROFILING
# ═══════════════════════════════════════════════════════════════════

print("\nCREATING COMPREHENSIVE DATA DICTIONARY")
print("═" * 70)

# ───────────────────────────────────────────────────────────────────
# TODO: Create detailed profile for each column
# Include: column name, type, null count, unique values, sample
# ───────────────────────────────────────────────────────────────────

def create_column_profile(df, dataset_name):
    """
    Create detailed profile for all columns in a dataset

    Parameters:
    -----------
    df : pd.DataFrame
        Dataset to profile
    dataset_name : str
        Name of dataset

    Returns:
    --------
    pd.DataFrame with column profiles
    """
    profiles = []

    for col in df.columns:
        # Get sample value (first non-null)
        sample = df[col].dropna().iloc[0] if df[col].count() > 0 else 'N/A'
        # Truncate long samples
        sample_str = str(sample)[:50]

        profile = {
            'Dataset': dataset_name,
            'Column_Name': col,
            'Data_Type': str(df[col].dtype),
            'Non_Null_Count': df[col].count(),
            'Null_Count': df[col].isna().sum(),
            'Null_%': round(df[col].isna().sum() / len(df) * 100, 2),
            'Unique_Values': df[col].nunique(),
            'Sample_Value': sample_str
        }

        profiles.append(profile)

    return pd.DataFrame(profiles)

# Create profiles for all datasets
print("Profiling all columns...")

all_profiles = []
for name, df in datasets.items():
    print(f"  ✓ Profiling {name}... ({len(df.columns)} columns)")
    profile_df = create_column_profile(df, name)
    all_profiles.append(profile_df)

# Combine all profiles
complete_dictionary = pd.concat(all_profiles, ignore_index=True)

print(f"\n✓ Created data dictionary with {len(complete_dictionary)} total columns")
print(f"  - Applicant: {len(applicant_df.columns)} columns")
print(f"  - SEVIS: {len(sevis_df.columns)} columns")
print(f"  - Connect: {len(connect_df.columns)} columns")
print(f"  - Outreach: {len(outreach_df.columns)} columns")

# Display sample from dictionary
print("\nData Dictionary Sample (first 10 rows):")
print("─" * 70)
print(complete_dictionary.head(10).to_string(index=False))
```

```python
# ═══════════════════════════════════════════════════════════════════
# SECTION 8: SUMMARY STATISTICS FOR KEY FIELDS
# ═══════════════════════════════════════════════════════════════════

print("\n\nKEY FIELD STATISTICS")
print("═" * 70)
print("⚠ IMPORTANT NOTE:")
print("  The percentages below are calculated from TOTAL RECORDS in each dataset.")
print("  Some datasets (especially Connect) contain DUPLICATE students (same student,")
print("  multiple records). True student-level rates will be calculated in Notebook 3")
print("  after proper deduplication.")
print("═" * 70)

# ───────────────────────────────────────────────────────────────────
# TODO: Show value distributions for important categorical fields
# Customize which fields to analyze based on business needs
# ───────────────────────────────────────────────────────────────────

# Applicant: Top Programs
print("\n1. TOP 10 PROGRAMS (Applicant Data)")
print("─" * 70)
if 'Major' in applicant_df.columns:
    program_counts = applicant_df['Major'].value_counts().head(10)
    for program, count in program_counts.items():
        pct = count / applicant_df['Major'].count() * 100
        # Truncate long program names
        program_short = program[:60] + '...' if len(program) > 60 else program
        print(f"  {program_short:60s}: {count:4d} ({pct:5.2f}%)")

# Applicant: Intake Terms
print("\n\n2. INTAKE TERM DISTRIBUTION (Applicant Data)")
print("─" * 70)
if 'Intake' in applicant_df.columns:
    intake_counts = applicant_df['Intake'].value_counts()
    for intake, count in intake_counts.items():
        pct = count / applicant_df['Intake'].count() * 100
        print(f"  {str(intake):20s}: {count:5d} ({pct:5.2f}%)")

# SEVIS: SEVIS Status
print("\n\n3. SEVIS STATUS (SEVIS Data)")
print("─" * 70)
if 'SEVIS_Status' in sevis_df.columns:
    status_counts = sevis_df['SEVIS_Status'].value_counts()
    for status, count in status_counts.items():
        pct = count / len(sevis_df) * 100
        print(f"  {str(status):20s}: {count:5d} ({pct:5.2f}%)")

# Connect: Deposit Status
print("\n\n4. DEPOSIT STATUS (Connect Data)")
print("─" * 70)
print("⚠ Based on ALL records (34,341) - includes duplicates")
print()
if 'Deposit_Status' in connect_df.columns:
    deposit_counts = connect_df['Deposit_Status'].value_counts()
    total_records = len(connect_df)

    for status, count in deposit_counts.items():
        pct = count / total_records * 100
        print(f"  {str(status):20s}: {count:6d} ({pct:5.2f}%)")

    # Store for later comparison
    deposit_yes_records = deposit_counts.get('Yes', 0)
    deposit_no_records = deposit_counts.get('No', 0)

# Connect: I-20 Status
print("\n\n5. I-20 STATUS (Connect Data)")
print("─" * 70)
print("⚠ Based on ALL records (34,341) - includes duplicates")
print()
if 'I_20_Status' in connect_df.columns:
    i20_counts = connect_df['I_20_Status'].value_counts()
    total_records = len(connect_df)

    for status, count in i20_counts.items():
        pct = count / total_records * 100
        print(f"  {str(status):20s}: {count:6d} ({pct:5.2f}%)")

    # Store for later comparison
    i20_yes_records = i20_counts.get('Yes', 0)
    i20_no_records = i20_counts.get('No', 0)

# Outreach: Student Status
print("\n\n6. STUDENT STATUS (Outreach Data)")
print("─" * 70)
print("✓ No duplicates in Outreach - these are unique students")
print()
if 'Student_Status' in outreach_df.columns:
    outreach_status = outreach_df['Student_Status'].value_counts()
    total_outreach = len(outreach_df)

    for status, count in outreach_status.items():
        pct = count / total_outreach * 100
        print(f"  {str(status):30s}: {count:4d} ({pct:5.2f}%)")

    # Store visa stats
    visa_approved = outreach_status.get('Visa approved', 0)
    visa_denied = outreach_status.get('Visa denied', 0)

# ═══════════════════════════════════════════════════════════════════
# CRITICAL CLARIFICATIONS & CONTEXT
# ═══════════════════════════════════════════════════════════════════

print("\n\n" + "═" * 70)
print("📊 UNDERSTANDING THE NUMBERS - CRITICAL CONTEXT")
print("═" * 70)

# ───────────────────────────────────────────────────────────────────
# Explain Connect Data duplication
# ───────────────────────────────────────────────────────────────────

print("\n1. CONNECT DATA DUPLICATION")
print("─" * 70)

unique_ref_ids = connect_df['Reference_ID'].nunique()
total_connect_records = len(connect_df)
duplicate_records = total_connect_records - unique_ref_ids

print(f"  Total Connect records:        {total_connect_records:,}")
print(f"  Unique students (Reference_ID): {unique_ref_ids:,}")
print(f"  Duplicate records:            {duplicate_records:,} ({duplicate_records/total_connect_records*100:.2f}%)")
print()
print("  💡 WHY? Connect appears to be an EVENT LOG, not a status table.")
print("     Each row may represent an interaction/update over time.")
print()
print("  ⚠ IMPLICATION: The percentages above use ALL records as denominator.")
print("     True student-level rates require deduplication (Notebook 3).")

# ───────────────────────────────────────────────────────────────────
# Show why percentages might match despite different denominators
# ───────────────────────────────────────────────────────────────────

print("\n\n2. DEPOSIT & I-20 RATES: Raw vs Student-Level")
print("─" * 70)
print("  The percentages shown above are based on:")
print(f"    Denominator: {total_connect_records:,} total records (with duplicates)")
print()
print("  Actual student-level calculation would use:")
print(f"    Denominator: {unique_ref_ids:,} unique students")
print()
print("  📊 COMPARISON TABLE:")
print("  " + "─" * 66)
print(f"  {'Metric':<30} {'All Records':<18} {'Unique Students*':<18}")
print("  " + "─" * 66)

# Calculate percentages both ways (estimate for unique students)
deposit_yes_pct_records = (deposit_yes_records / total_connect_records * 100)
deposit_yes_pct_students = 21.9  # From our previous analysis

i20_yes_pct_records = (i20_yes_records / total_connect_records * 100)
i20_yes_pct_students = 34.7  # From our previous analysis

print(f"  {'Deposit = Yes %':<30} {deposit_yes_pct_records:>6.2f}% ({deposit_yes_records:>6,}) {i20_yes_pct_students:>6.1f}% (~1,512)")
print(f"  {'I-20 = Yes %':<30} {i20_yes_pct_records:>6.2f}% ({i20_yes_records:>6,}) {deposit_yes_pct_students:>6.1f}% (~2,390)")
print("  " + "─" * 66)
print("  *Estimated from deduplication analysis (see Notebook 3)")
print()
print("  💡 INSIGHT: Percentages are SIMILAR because most students maintain")
print("     the same status across their multiple records. But the COUNTS differ!")

# ───────────────────────────────────────────────────────────────────
# Clarify visa approval rate misconception
# ───────────────────────────────────────────────────────────────────

print("\n\n3. VISA APPROVAL RATE - THE MISLEADING 5.5%")
print("─" * 70)

total_applicants = len(applicant_df)
visa_pct_of_applicants = (visa_approved / total_applicants) * 100

print(f"  Visa approved (Outreach):     {visa_approved} students")
print(f"  Total applicants:             {total_applicants:,}")
print()
print(f"  ❌ WRONG: {visa_approved} / {total_applicants:,} = {visa_pct_of_applicants:.2f}%")
print("     (Comparing Outreach subset to ALL applicants)")
print()

if visa_denied > 0:
    total_decided = visa_approved + visa_denied
    true_approval_rate = (visa_approved / total_decided) * 100
    completion_rate = (total_decided / total_outreach) * 100

    print(f"  ✓ CORRECT Approval Rate:")
    print(f"    {visa_approved} approved / ({visa_approved} approved + {visa_denied} denied)")
    print(f"    = {visa_approved} / {total_decided} = {true_approval_rate:.1f}%")
    print()
    print(f"  ✓ Visa Application Completion Rate:")
    print(f"    {total_decided} decided / {total_outreach} in outreach = {completion_rate:.1f}%")
    print()
    print("  💡 INSIGHT: Among students who applied for visa and got a decision,")
    print(f"     the approval rate is {true_approval_rate:.1f}% - much higher than the misleading 5.5%!")

```
